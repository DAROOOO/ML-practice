```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns
```

### 여러열을 특정갯수로 축소하는 알고리즘 PCA

```python
iris = datasets.load_iris()
iris_x = iris.data
iris_y = iris.target

# 4개의 열을 1개로 차원축소
model = decomposition.PCA(n_components=1)
model.fit(iris_x)
iris_x1 = model.transform(iris_x)
```

```python
print(iris_x.shape)
print(iris_y.shape)
```

```python
# 1차원 데이터를 활용해 커널밀도함수 시각화 작성.
# 커널 밀도 함수는 일정 범위안의 속할 확률을 나타낸다.
# sns.distplot(boston_x1[(boston_y<=40) and (boston_y >=20)], color="g", bins=20, kde=False)
import seaborn as sns

# bins : Specification of hist bins, or None to use Freedman-Diaconis rule.
# kde : Whether to plot a gaussian kernel density estimate
sns.distplot(iris_x1[iris_y==0], color="b", bins=20, kde=False)
sns.distplot(iris_x1[iris_y==1], color="g", bins=20, kde=False)
sns.distplot(iris_x1[iris_y==2], color="r", bins=20, kde=False)

plt.xlim(-6, 6)

plt.show()
```

```python
model = decomposition.PCA(n_components=3)
model.fit(iris_x)
iris_x = model.fit_transform(iris_x)
# PCA plot of 2 PCs
plt.scatter(iris_x[:, 0], iris_x[:, 1], c=iris.target)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
# PCA plot of 3 PCs
fig = plt.figure()
ax = Axes3D(fig, elev=48, azim=134) # Set the elevation and azimuth of the axes. (축의 고도와 방위각)

ax.scatter(iris_x[:, 0], iris_x[:, 1], iris_x[:, 2], c=iris.target, edgecolor='w', s=100)
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.dist = 12 # 값이 커지면 전체 plot 이 작아짐

plt.show()
```
